```python
# Importing the pandas library for data manipulation and analysis
import pandas as pd

# Importing the numpy library for numerical computing and array operations
import numpy as np

# Importing the matplotlib.pyplot module for creating static, animated, and interactive visualizations
import matplotlib.pyplot as plt

# Importing the seaborn library for statistical data visualization
import seaborn as sns

```

```python
# Load the Excel file into a DataFrame
df=pd.read_excel("/content/datset.xlsx")

# Display the DataFrame
df
```

```python
# Retrieve the dimensions of the DataFrame: (number of rows, number of columns)
df.shape
```

```python
# Count duplicate rows
df.duplicated().sum()
```

```python
# Count and sort the number of missing (NaN) values in each column in descending order
df.isnull().sum().sort_values(ascending=False)
```

```python
# Iterate through each column in the DataFrame to display its data type, sorted unique values, and the unique values themselves

for i in df:
    print(i, ": ", df[i].dtypes)              # Print the data type of each column
    print(df[i].sort_values().unique())       # Print sorted unique values of each column
    print()                                   # Print a newline for better readability
```

```python
# Filter rows with at least one missing (NaN) value across any column

null_rows = df[df.isnull().any(axis=1)]
null_rows
```

```python
# Fill missing values: use mode for categorical columns and median for numerical columns

for i in df:
    if df[i].dtypes == "object":
        df[i].fillna(df[i].mode()[0], inplace=True)
    else:
        df[i].fillna(df[i].median(), inplace=True)
```

```python
# Count the number of missing (NaN) values in each column of the DataFrame
df[i].isnull().sum()
```

```python
# Generate summary statistics for numerical columns in the DataFrame
df.describe()
```

#Sum of sales by region

```python
# Group by 'Region', sum 'TotalSales', and reset index to convert 'Region' from index to column
total_sales=df.groupby("Region")["TotalSales"].sum().sort_values(ascending=False).reset_index()
print(total_sales)
print()
# Plotting with Seaborn
plt.figure(figsize=(6, 3))
sns.barplot(data=total_sales, x='Region', y='TotalSales', hue="Region")
plt.title('TotalSales by region')
plt.xlabel('Region')
plt.ylabel('TotalSales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The North region is the best-performing in terms of sales.

East has room for improvement and might benefit from targeted marketing or operational strategies.

#Average sales per product

```python
# Calculate average 'TotalSales' per 'Product' and reset index to flatten the DataFrame
avg_sales = df.groupby("Product")["TotalSales"].mean().sort_values(ascending=False).reset_index()
print(avg_sales)
print()
# Plotting with Seaborn
plt.figure(figsize=(6, 3))
sns.barplot(data=avg_sales, x='Product', y='TotalSales', hue="Product")
plt.title('Average TotalSales per Product')
plt.xlabel('Product')
plt.ylabel('Average TotalSales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Focusing on **Laptop** could maximize revenue due to its strong average sales.

**Tablet** maintains a healthy sales average, but there's room to explore strategies to boost its performance further.

**Smart Watch** may require targeted promotions or product enhancements to increase its market appeal.

#Highest & lowest selling products

```python
# Lowest selling products
product_counts[product_counts==product_counts.min()].reset_index()
```

# Compute mean, median, and standard deviation for numerical columns using NumPy


```python
# Iterate over each column in the DataFrame
for i in df:
    # Check if the column is numeric
    if np.issubdtype(df[i].dtype, np.number):
        print(i, "mean:  ", np.mean(df[i]))
        print("      median:  ", np.median(df[i]))
        print("        std:  ", np.std(df[i]))
        print()
```
